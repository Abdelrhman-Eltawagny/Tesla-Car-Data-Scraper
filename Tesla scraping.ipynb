```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
from selenium import webdriver
```

```python
links = []
page_num = 1

while True:
    page = requests.get(f'https://www.cars.com/shopping/tesla/?page={page_num}')
    src = page.content
    soup = BeautifulSoup(src , 'lxml')
    car_links = soup.find('div' , {'data-tracking-orientation':'vertical'}).find_all('div',{'class':'vehicle-card'})
    for i in range(len(car_links)):
        links.append('https://www.cars.com' + car_links[i].find('a').attrs['href'])
    
    print(f'page {page_num} : Done')
    page_num += 1
    if page_num == 24:
        print('done all')
        break
```

```python
model_name = []
model_year = []
salaries = []
mile_galon = []
exterior_color= []
interior_color = []
drivetrain = []
fuel_type = []
transmission = []
engine_type = []
```

```python
for link in links:
    page = requests.get(link)
    content = page.content
    soup = BeautifulSoup(content, features='html.parser')
    
    #get name and year of tesla model
    model = soup.find('h1' , class_ = 'listing-title').text.split(' Tesla Model ')
    model_year.append(model[0])
    model_name.append(model[1])
    
    #get salary of car
    salary = soup.find('div', {'class' : 'price-section'})
    salaries.append(salary.contents[1].text[1:].replace(',' , ''))
    
    #get mile/galon of car
    milgal = soup.find('div' , class_ = 'listing-mileage')
    mile_galon.append(milgal.text.replace('mi.' , '').replace(',' , '').strip())
    
    #Get the exterior color of the car
    basics = soup.find('dl' , class_ = 'fancy-description-list')
    exterior_color.append(basics.contents[3].text.strip())
    
    #Get the exterior color of the car
    interior_color.append(basics.contents[7].text.strip())
    
    #Get drivetrain of the cat
    drivetrain.append(basics.contents[11].text.strip())
    
    #Get fuel type of the cat
    fuel_type.append(basics.contents[15].text.strip())
    
    #Get transmission of the cat
    transmission.append(basics.contents[19].text.strip())
    
    #Get engine type of the cat
    engine_type.append(basics.contents[23].text.strip())
```

```python
Tesla_model = pd.DataFrame({'Model Name' : model_name , 'Model Year' : model_year, 'Mile/Galon' : mile_galon
                            , 'Exterior Color' : exterior_color ,'Interior Color' :interior_color ,'Drivetrain' : drivetrain
                            ,'Fuel Type' : fuel_type ,'Transmission' :transmission ,'Engine Type' : engine_type,'Price' : salaries})
Tesla_model.head()
```
